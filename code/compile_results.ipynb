```python
import numpy as np
import xarray as xr

file_path = '/scratch/rd53/wy2165/AABR/'
glacier = xr.open_dataset(file_path + 'glacier_statistics.nc')
pygem   = xr.open_dataset(file_path + 'results_AAR_debriscalving.nc')
oggm    = xr.open_dataset(file_path + 'results_oggm.nc')

rgi_ids = glacier['RGIId']
results = oggm.sel(RGIId=rgi_ids)
results = results.merge(glacier)

pygem = pygem.set_coords('RGIId')
pygem = pygem.swap_dims({'glac': 'RGIId'})
pygem = pygem.drop_vars('glac')
pygem = pygem.sel(RGIId=rgi_ids)
pygem = pygem[['is_debris', 'is_icecap', 'is_tidewater', 'glac_prec','glac_temp','glac_acc','glac_refreeze','glac_melt','glac_frontalablation','glac_massbaltotal']]

results = results.merge(pygem)
loc = np.where(np.isnan(results['AABR_median_compile'].values))[0]
results['AAR_median_compile'].values[loc] = np.nan
results['THAR_median_compile'].values[loc] = np.nan
results['ELA_median_compile'].values[loc] = np.nan
loc = np.where(np.isnan(results['AABR_median_intercept'].values))[0]
results['AAR_median_intercept'].values[loc] = np.nan
results['THAR_median_intercept'].values[loc] = np.nan
results['ELA_median_intercept'].values[loc] = np.nan
loc = np.where(np.isnan(results['AABR_median_steady'].values))[0]
results['AAR_median_steady'].values[loc] = np.nan
results['THAR_median_steady'].values[loc] = np.nan
results['ELA_median_steady'].values[loc] = np.nan

loc = np.where(np.isnan(results['AABR_min_compile'].values))[0]
results['AAR_min_compile'].values[loc] = np.nan
results['THAR_min_compile'].values[loc] = np.nan
results['ELA_min_compile'].values[loc] = np.nan
loc = np.where(np.isnan(results['AABR_min_intercept'].values))[0]
results['AAR_min_intercept'].values[loc] = np.nan
results['THAR_min_intercept'].values[loc] = np.nan
results['ELA_min_intercept'].values[loc] = np.nan
loc = np.where(np.isnan(results['AABR_min_steady'].values))[0]
results['AAR_min_steady'].values[loc] = np.nan
results['THAR_min_steady'].values[loc] = np.nan
results['ELA_min_steady'].values[loc] = np.nan

loc = np.where(np.isnan(results['AABR_max_compile'].values))[0]
results['AAR_max_compile'].values[loc] = np.nan
results['THAR_max_compile'].values[loc] = np.nan
results['ELA_max_compile'].values[loc] = np.nan
loc = np.where(np.isnan(results['AABR_max_intercept'].values))[0]
results['AAR_max_intercept'].values[loc] = np.nan
results['THAR_max_intercept'].values[loc] = np.nan
results['ELA_max_intercept'].values[loc] = np.nan
loc = np.where(np.isnan(results['AABR_max_steady'].values))[0]
results['AAR_max_steady'].values[loc] = np.nan
results['THAR_max_steady'].values[loc] = np.nan
results['ELA_max_steady'].values[loc] = np.nan

results.to_netcdf(file_path + 'results_all_debriscalving.nc')
```
